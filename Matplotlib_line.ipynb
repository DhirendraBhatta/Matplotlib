# Line Graphs

```python
x=[1,2,3,5,6]
y=[3,3,7,1,3]
```

```python
import numpy as np
# color = k-black, b-blue , r-red, g-green, m-magenta, c-cyan, y-yellow
plt.plot(x,y,linewidth =5, color='m')
plt.title('Test')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.xticks(np.arange(1,6.5,.5))#dividing x axis in intervals(stating point,ending point-exclusive,interval)
# plt.yticks(np.arange(1,8, .5))
plt.grid()
plt.show()
```
