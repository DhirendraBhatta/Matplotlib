```python
from matplotlib import pyplot as plt
```

```python
from matplotlib import style
```

```python
x = [1,3,4]
y = [2,4,4]
x1 = [1,2,4]
y1 = [2,2,4]
```

```python
import numpy as np
plt.plot(x, y, linewidth=5, color="r", label="Gold Price")
plt.plot(x1, y1, linewidth=5, color='g', label="Silver Price")
plt.legend()
plt.xticks(ticks =np.arange(1,8),labels=['Sun','Mon','Tue',"Wed",'Thurs','Fri','Sat'])
plt.title('Gold Vs Silver Price Comparision')
plt.xlabel('Day', fontsize=15)
plt.ylabel('Price in $',fontsize=15)
plt.grid(True,color='y')
plt.savefig('linegraph1.png')
plt.savefig('linegraph2.jpg')
plt.savefig('linegraph3.pdf')
plt.show()
```
