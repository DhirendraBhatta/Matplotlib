```python
from matplotlib import pyplot as plt
```

```python
from matplotlib import style
```

```python
x=[1,2,3,4,5,6]
y=[2,3,4,5,6,7]
```

```python
import numpy as np
plt.plot(x,y,linewidth=5 ,color='g')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("Line Graph")
plt.xticks(np.arange(1,6.5, .5))
plt.yticks(np.arange(2,7.5,.5))
plt.grid()
plt.show()
```
