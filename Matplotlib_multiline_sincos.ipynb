```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0, 4*np.pi, 0.1)    #4*3.14(pi)
y = np.sin(x)
z = np.cos(x)


plt.plot(x, y, "g", linewidth=5 ,label='sin curve')
plt.plot(x, z, 'r' , linewidth=5, label='cos curve')
plt.legend()
plt.tight_layout()
plt.savefig("multiline.png")
plt.show()
```
