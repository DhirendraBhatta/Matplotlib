# Line Graphs

```python
x = [1,2,3,4,5,6]
y = [2,3,4,5,6,7]
```

```python
import numpy as np
plt.plot(x,y, linewidth=5, color='g')
plt.title("Test")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.xticks(np.arange(1,6.5, .5))
plt.grid()
plt.show()
```

```python
x = [1,3,4]
y = [2,4,4]
x1 = [1,2,4]
y1 = [2,2,4]

plt.plot(x, y, "r", linewidth=5, label="Line 1")
plt.plot(x1, y1, "g", linewidth=5, label="Line 2")
plt.legend()
plt.grid(True, color="c")
plt.show()
```

# Bar Graphs

```python
plt.bar([2,4],[1,2], label="Example 1", color="g")
plt.bar([1,3],[1,2], label="Example 2", color="k")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.show()
```

```python
import numpy as np
label = ['Adventure', 'Action', 'Drama', 'Comedy', 'Thriller/Suspense']
no_movies = [
    941,
    854,
    4595,
    2125,
    942
]
# index = np.arange(len(label))
plt.bar(label, no_movies)
plt.xlabel('Genre', fontsize=5)
plt.ylabel('No of Movies', fontsize=5)
plt.xticks(fontsize=10, rotation=30)
plt.title('Market Share for Each Genre 1995-2017')
plt.show()
```

# Histogram

```python
import numpy as np
population_bins = np.arange(-20,131,10)
population_sample = np.random.normal(40, 14, 10000)
# print(population_sample)
plt.hist(population_sample, population_bins, rwidth=0.8)
# plt.hist(population_sample,population_bins, histtype='bar', rwidth=0.8)
# plt.xticks(population_bins)
plt.show()
```

# Pie Chart

```python
slices = [7,2,2,13]
divs = ["education", "food", "house rent", "daily needs"]
colors = ["c", "m", "r", "b"]
plt.pie(slices, labels=divs, colors=colors, shadow=True, autopct='%1.2f%%')
plt.legend()
plt.show()
```

```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)

t1 = np.arange(0, 5, 0.1)
t2 = np.arange(0, 5, 0.02)

plt.subplot(121)
plt.plot(t1, f(t1), 'bo', t2, f(t2))

plt.subplot(122)
plt.plot(t2, np.cos(2*np.pi*t2))

plt.show()
```
